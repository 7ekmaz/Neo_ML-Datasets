# Memory

we will determine the (SDD,HDD,Flash,Hybrid) values

```python

# Function to extract storage information
def extract_storage_info(row):
    storage_info = {
        'SSD': 0,
        'HDD': 0,
        'Flash': 0,
        'Hybrid': 0
    }

    # Replace 'Flash Storage' with 'Flash' for consistency
    row = row.replace('Flash Storage', 'Flash')

    # Define patterns
    patterns = {
        'SSD': r'(\d+(?:\.\d+)?)(TB|GB) SSD',
        'HDD': r'(\d+(?:\.\d+)?)(TB|GB) HDD',
        'Flash': r'(\d+(?:\.\d+)?)(TB|GB) Flash',
        'Hybrid': r'(\d+(?:\.\d+)?)(TB|GB) Hybrid'
    }

    for key, pattern in patterns.items():
        matches = re.findall(pattern, row)
        for match in matches:
            size, unit = match
            size = float(size.strip())
            if 'TB' in unit:
                size *= 1000  # Convert TB to GB
            storage_info[key] += size

    return pd.Series(storage_info)

# Apply the function to each row to create new columns
data[['SSD', 'HDD', 'Flash', 'Hybrid']] = data['Memory'].apply(extract_storage_info)

print("\nDataFrame after extracting storage information:")
data.head()
```

# Random Forest

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
model=RandomForestRegressor()

param_grid={
    'n_estimators':[10,20,50,150,100,200,250,300], 'max_depth': [4,5,3,7]
} 

grid_search=GridSearchCV(model,param_grid=param_grid)
grid_search.fit(train_x,train_y)


 
print(grid_search.best_params_)


model=RandomForestRegressor(**grid_search.best_params_)
model.fit(train_x,train_y)


y_pred=model.predict(test_x)
print("R2 SCORE: ",r2_score(test_y,y_pred))

y_pred_train=model.predict(train_x)
print("R2 SCORE: ",r2_score(train_y,y_pred_train)) #For overfitting check
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV


rf = RandomForestRegressor(n_estimators=160, random_state=33)

# Define the parameter grid
param_grid_regressor = {
    'n_estimators': [10, 20, 50, 150, 100, 200, 300],               # Number of trees in the forest
    'max_depth': [3, 4, 5, 7, 10],           # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],                    # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],                      # Minimum number of samples required to be at a leaf node
    'bootstrap': [True, False],                         # Whether bootstrap samples are used when building trees
    'n_jobs': [-1]                                     # Number of jobs to run in parallel (-1 means using all processors)
}
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid_regressor, cv=5, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(train_x, train_y)

# Get the best parameters
best_params = grid_search.best_params_
print("Best parameters found: ", best_params)

# Get the best model
best_rf = grid_search.best_estimator_

# Predict on test data
y_pred = best_rf.predict(test_x)
y_pred_train = best_rf.predict(train_x)



```

```python
mse = mean_squared_error(test_y, y_pred)
r2 = r2_score(test_y, y_pred)
print(f"\nMean Squared Error for Prediction: {mse:.2f}")
print(f"R2 Score for Prediction: {r2:.9f} \n\n\n\n")


mse = mean_squared_error(train_y, y_pred_train)
r2 = r2_score(train_y, y_pred_train)
print(f"\nMean Squared Error for training: {mse:.2f}")
print(f"R2 Score for training: {r2:.9f} ")
```
